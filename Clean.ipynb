```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV and JSON's are located
education_csv = "Education.csv"
education_json = "json_education.json"
price_csv = "Housing_Price.csv"
price_json = "json_housing_price.json"

# Import the data into a Pandas DataFrame
education1_df = pd.read_csv(education_csv)
education2_df = pd.read_json(education_json)
price1_df = pd.read_csv(price_csv)
price2_df = pd.read_json(price_json)
#education1_df
#education2_df
#price1_df
#price2_df
```

```python
merge_csv_df = pd.merge(education1_df, price1_df, on="state", how="outer")
merge_csv_df["MedianValue"] = merge_csv_df["MedianValue"].astype(float).map("${:,.2f}".format)
merge_csv_df["medianValueZillow"] = merge_csv_df["medianValueZillow"].astype(float).map("${:,.2f}".format)
merge_csv_df
```

```python
merge_json_df = pd.merge(education2_df, price2_df, on="state", how="outer")
merge_json_df["MedianValue"] = merge_json_df["MedianValue"].astype(float).map("${:,.2f}".format)
merge_json_df["medianValueZillow"] = merge_json_df["medianValueZillow"].astype(float).map("${:,.2f}".format)
merge_json_df
```
